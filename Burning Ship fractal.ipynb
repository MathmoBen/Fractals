# Burning Ship fractal

In this notebook we tweak the construction of the Mandelbrot set to construct the Burning Ship fractal. Details here: https://en.wikipedia.org/wiki/Burning_Ship_fractal#External_links

```python
# The usual imports

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from dataclasses import dataclass
from math import log
```

```python
# Another useful function

def complex_matrix(xmin, xmax, ymin, ymax, pixel_density):
    re = np.linspace(xmin, xmax, int((xmax - xmin) * pixel_density))
    im = np.linspace(ymin, ymax, int((ymax - ymin) * pixel_density))
    return re[np.newaxis, :] + im[:, np.newaxis] * 1j
```

```python
# A function for gauging whether a point is diverging or not

def is_stable(c, num_iterations):
    z = 0
    for _ in range(num_iterations):
        z = (abs(z.real) + abs(z.imag)*1j) ** 2 + c
    return abs(z) <= 2
```

```python
def get_members(c, num_iterations):
    mask = is_stable(c, num_iterations)
    return c[mask]
```

```python
#A slightly more sophisticated approach to the same thing

c = complex_matrix(-2, 1, -2, 1, pixel_density=512)
plt.imshow(is_stable(c, num_iterations=20), cmap="binary")
plt.gca().set_aspect("equal")
plt.axis("off")
plt.tight_layout()
plt.show()
```
